# Matplotlib wala bar graph

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25) # Random
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25) # FIFO
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25) # LRU
ax.set_xticks(X + 0.25)
ax.set_xticklabels(['4 bytes', '8 bytes', '16 bytes', '32 bytes'])
```

```python
# Random, FIFO, LRU
data = [[0.457, 0.675, 0.733, 0.767, 0.782],
[0.496, 0.738, 0.747, 0.777,0.796],
[0.447, 0.758, 0.758, 0.782, 0.797]]


X = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1.5,1.5])
ax.set_title('Cache Hit Ratio Vs Cache size with respect to different replacement algorithms')
ax.set_ylabel('Hit Ratio')
ax.set_xlabel('Byte wise for Random, FIFO and LRU')
ax.set_ylim(0,1)
ax.bar(X + 0.00, data[0], color = 'm', width = 0.25) # Random
ax.bar(X + 0.25, data[1], color = 'b', width = 0.25) # FIFO
ax.bar(X + 0.50, data[2], color = 'c', width = 0.25) # LRU
ax.set_xticks(X + 0.25)
ax.set_xticklabels(['4 bytes', '8 bytes', '16 bytes', '32 bytes', '64 bytes'])
ax.legend(labels = ('Random', 'FIFO', 'LRU'), loc = 'upper right')
fig.savefig('One.png')
```
